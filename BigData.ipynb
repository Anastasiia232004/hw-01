```python
from collections import defaultdict
from multiprocessing import Pool
import re

# Функція Map: розбиває текст на слова та повертає пари (слово, 1)
def map_function(chunk):
    word_count = defaultdict(int)
    words = re.findall(r'\w+', chunk.lower())
    for word in words:
        word_count[word] += 1
    return word_count

# Функція Reduce: об'єднує результати з Map і сумує значення
def reduce_function(mapped_data):
    combined_data = defaultdict(int)
    for partial_result in mapped_data:
        for word, count in partial_result.items():
            combined_data[word] += count
    return combined_data

# Функція для розбивки файлу на частини
def chunk_file(file_path, chunk_size=1024):
    with open(file_path, 'r', encoding='utf-8') as file:
        while chunk := file.read(chunk_size):
            yield chunk

# Основна функція MapReduce
def map_reduce(file_path, chunk_size=1024, num_workers=4):
    # Розбивка файлу на частини
    chunks = list(chunk_file(file_path, chunk_size))

    # Використання пулу процесів для Map
    with Pool(num_workers) as pool:
        mapped_results = pool.map(map_function, chunks)

    # Reduce
    final_result = reduce_function(mapped_results)
    return final_result

# Завантаження файлу в середовище Colab
from google.colab import files

uploaded = files.upload()  # Завантажте файл у середовище Colab
file_path = list(uploaded.keys())[0]  # Отримання назви завантаженого файлу

# Виконання програми
if __name__ == '__main__':
    word_count = map_reduce(file_path, chunk_size=1024, num_workers=4)

    # Виведення результату
    for word, count in sorted(word_count.items(), key=lambda x: x[1], reverse=True):
        print(f'{word}: {count}')
```

```python
from collections import defaultdict
from multiprocessing import Pool
import re

# Функція Map: аналізує лог-файл, повертає пари (IP-адреса, 1)
def map_function(log_chunk):
    ip_count = defaultdict(int)
    log_lines = log_chunk.strip().split("\n")
    for line in log_lines:
        match = re.match(r'(\d+\.\d+\.\d+\.\d+)', line)
        if match:
            ip_address = match.group(1)
            ip_count[ip_address] += 1
    return ip_count

# Функція Reduce: об'єднує результати з Map і сумує кількість запитів
def reduce_function(mapped_data):
    combined_data = defaultdict(int)
    for partial_result in mapped_data:
        for ip, count in partial_result.items():
            combined_data[ip] += count
    return combined_data

# Функція для розбивки лог-файлу на частини
def chunk_file(file_path, chunk_size=1024):
    with open(file_path, 'r', encoding='utf-8') as file:
        while chunk := file.read(chunk_size):
            yield chunk

# Основна функція MapReduce
def map_reduce(file_path, chunk_size=1024, num_workers=4):
    # Розбивка файлу на частини
    chunks = list(chunk_file(file_path, chunk_size))

    # Використання пулу процесів для Map
    with Pool(num_workers) as pool:
        mapped_results = pool.map(map_function, chunks)

    # Reduce
    final_result = reduce_function(mapped_results)
    return final_result

# Завантаження файлу в середовище Colab (якщо працюєте в Google Colab)
from google.colab import files

# Завантажити файл
uploaded = files.upload()
file_path = list(uploaded.keys())[0]

# Виконання програми
if __name__ == '__main__':
    # Задайте шлях до файлу
    word_count = map_reduce(file_path, chunk_size=1024, num_workers=4)

    # Виведення результату
    for ip, count in sorted(word_count.items(), key=lambda x: x[1], reverse=True):
        print(f'{ip}: {count}')
```
